```python
df1 = df[['content']].copy()
```

### REMOVING NON TEXT VALUES

```python
import re
def clean(text):
    text = re.sub('[^A-Za-z]+', ' ', text)
    return text

df1['Cleaned'] = df1['content'].apply(clean)
df1.head()
```

### NORMALIZATION

```python
df1['Lowercase'] = df1['Cleaned'].apply(lambda x: " ".join(word.lower() for word in x.split()))
df1.head(2)
```

### REMOVING STOPWORDS

```python
nltk.download('stopwords')
```

```python
from nltk.corpus import stopwords
stop_words = stopwords.words('english')
print(stop_words)
```

```python
df1.head()
```

```python
df1['stopwords'] = df1['Lowercase'].apply(lambda x: " ".join(word for word in x.split() if word not in stop_words))
```

```python
df1.head(2)
```

```python
from nltk.tokenize import word_tokenize
from nltk.corpus import wordnet
from nltk import pos_tag
nltk.download('punkt_tab')
nltk.download('punkt')
nltk.download('wordnet')
nltk.download('averaged_perceptron_tagger_eng')
```

### PARTS OF SPEECH DICTIONARY

```python
pos_dict = {'J': wordnet.ADJ, 'V': wordnet.VERB, 'N': wordnet.NOUN, 'R': wordnet.ADV}

def token_pos(text):
    return [
        (word, pos_dict.get(tag[0]))
        for word, tag in pos_tag(word_tokenize(text))
    ]

df1['POS tagged'] = df1['stopwords'].apply(token_pos)
```

```python
df1.head(2)
```

### LEMATATION

```python
from nltk.stem import WordNetLemmatizer
wordnet_lemmatizer = WordNetLemmatizer()

def lemmatize(pos_data):
    ans = " "
    for word, pos in pos_data:
        if not pos:
            lemma = word
        else:
            lemma = wordnet_lemmatizer.lemmatize(word, pos=pos)
        ans = ans + " " + lemma
    return ans

df1['Lemma'] = df1['POS tagged'].apply(lemmatize)
```

```python
def scores(RATES):
    if RATES > 3:  # Positive threshold
        return 0   # positive
    elif RATES < 3:  # Negative threshold
        return 1 # negative
    else:
        return 2 # neutral
df["scores"]=df["score"].apply(scores)
```

```python
X=df1["Lemma"]
Y=df["scores"]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
```

```python
X_train.shape
```

# TF_IDF

```python
from sklearn.feature_extraction.text import TfidfVectorizer
vectorizer = TfidfVectorizer(
    ngram_range=(1,2),
    stop_words='english',
    min_df=3,
    sublinear_tf=True
)
```

```python
X_train_vec = vectorizer.fit_transform(X_train)
X_test_vec  = vectorizer.transform(X_test)
```

```python
from sklearn.naive_bayes import MultinomialNB
model = MultinomialNB()
model.fit(X_train_vec, Y_train)
```

```python
Y_pred = model.predict(X_test_vec)
```

```python
from sklearn.metrics import accuracy_score
print(accuracy_score(Y_test,Y_pred))
```

```python
def predict_review(review):
    review_vec = vectorizer.transform([review])
    prediction = model.predict(review_vec)[0]

    label_map = {
        0: "positive",
        1: "negative",
        2: "positive"
    }
    return label_map[prediction]
print(predict_review("not liked"))
```
